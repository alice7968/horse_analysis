### race_data 数値に変換
- surface(馬場) "ダ": 0, "芝": 1
- weather(天候) "晴": 0, "曇": 1, "小雨": 2, "雨": 3, "小雪": 4, "雪": 5
- ground_status(馬場状態) "良": 0, "稍重": 1, "重": 2 , "不良": 3

```python
horse_df = pd.read_csv(PARENTDIR + "/final_cleaned_horse_data.csv")
horse_df = horse_df.drop(columns=["sex_and_age", "owner_id"])
cols = ["horse_id", "rider_id", "tamer_id"]
for col in cols:
    if col in horse_df.columns:
        horse_df[col] = horse_df[col].astype("int64").astype(str).str.zfill(5)
horse_df["horse_weight_dif"] = horse_df["horse_weight_dif"].fillna(0)
horse_df.head(5)
```

```python
import pandas as pd
horse_df["goal_time_dif"] = pd.NA

# goal_time_dif を計算する関数
def calculate_goal_time_dif(group):
    group = group.copy()
    group = group.sort_values("rank")
    
    rank1_time = group.loc[group["rank"] == 1, "goal_time"]
    rank2_time = group.loc[group["rank"] == 2, "goal_time"]
    
    if not rank1_time.empty:
        rank1_time = rank1_time.values[0]
        group["goal_time_dif"] = group["goal_time"] - rank1_time
        if not rank2_time.empty:
            rank2_time = rank2_time.values[0]
            group.loc[group["rank"] == 1, "goal_time_dif"] = rank1_time - rank2_time
    
    return group

# race_idごとに処理を適用
horse_df = horse_df.groupby("race_id", group_keys=False).apply(calculate_goal_time_dif).reset_index(drop=True)

# 結果の確認（必要に応じて）
horse_df[["race_id", "rank", "goal_time", "goal_time_dif"]].head()
```

### horseとraceをmerge!

```python
# マージ前に fillna してもよいが、型を維持するならマージ後に行う
merge_df = pd.merge(horse_df, race_df, how="left", on="race_id")

# NaN を -1 で埋めてから Int64（nullable int）に変換
int_cols = ["race_number", "surface", "weather", "ground_status","headcount", "frame_number_first", "horse_number_first", "frame_number_second", 
            "horse_number_second", "frame_number_third", "horse_number_third", "is_obstacle", "race_course_id"]  
for col in int_cols:
    merge_df[col] = pd.to_numeric(merge_df[col], errors="coerce").fillna(-1).astype("Int64")
merge_df.head(3)
```

```python
merge_df[["last_time","is_obstacle","direction"]].head(10)
```

```python
merge_df.loc[merge_df["direction"] == -1, "is_obstacle"] = 1
merge_df[["last_time","is_obstacle","direction"]].head(10)
```

```python
# 標準化されたhalf_way_difを計算
merge_df["standard_half_way_dif"] = merge_df["half_way_dif"] / merge_df["headcount"]

# horse_type を初期化（デフォルトは 0）
merge_df["horse_type"] = 0

# 条件を満たす行のみを対象にする
valid_idx = merge_df["goal_time_dif"] < 1

# 逃げ: (half_way_dif + rank) <= 2 and half_way_dif <= 1
cond_nige = (
    valid_idx &
    (merge_df["half_way_dif"] + merge_df["rank"] <= 2) &
    (merge_df["half_way_dif"] <= 1)
)
merge_df.loc[cond_nige, "horse_type"] = 1

# 先行: 0.1 < std <= 0.4
cond_senko = (
    valid_idx &
    (merge_df["standard_half_way_dif"] > 0.1) &
    (merge_df["standard_half_way_dif"] <= 0.4)
)
merge_df.loc[cond_senko, "horse_type"] = 2

# 差し: 0.4 < std <= 0.7
cond_sashi = (
    valid_idx &
    (merge_df["standard_half_way_dif"] > 0.4) &
    (merge_df["standard_half_way_dif"] <= 0.7)
)
merge_df.loc[cond_sashi, "horse_type"] = 3

# 追い込み: 0.7 < std <= 1.0
cond_oikomi = (
    valid_idx &
    (merge_df["standard_half_way_dif"] > 0.7) &
    (merge_df["standard_half_way_dif"] <= 1.0)
)
merge_df.loc[cond_oikomi, "horse_type"] = 4
merge_df[["race_id", "rank", "goal_time", "goal_time_dif", "horse_type"]].head(10)
```

```python
# 標準化された馬番
merge_df["standard_horse_number"] = merge_df["horse_number"] / merge_df["headcount"]

# 初期値を 0 で埋める
merge_df["frame_type"] = 0

# 条件に応じて分類
merge_df.loc[
    (merge_df["standard_horse_number"] >= 0.0) & (merge_df["standard_horse_number"] < 0.25),
    "frame_type"
] = 1  # 内枠

merge_df.loc[
    (merge_df["standard_horse_number"] >= 0.25) & (merge_df["standard_horse_number"] < 0.5),
    "frame_type"
] = 2  # 中内枠

merge_df.loc[
    (merge_df["standard_horse_number"] >= 0.5) & (merge_df["standard_horse_number"] < 0.75),
    "frame_type"
] = 3  # 中外枠

merge_df.loc[
    (merge_df["standard_horse_number"] >= 0.75) & (merge_df["standard_horse_number"] <= 1.0),
    "frame_type"
] = 4  # 外枠
```

```python
import pandas as pd
import os
parent_dir = os.path.dirname(os.getcwd())

# horse_type=0は無視、分析対象から除外
filtered_df = merge_df[merge_df["horse_type"] > 0].copy()

# 日付・horse_idでソート（未来情報排除のため）
filtered_df = filtered_df.sort_values(["horse_id", "date"])

# analy_horse_typeを入れるリスト
analy_horse_types = []

# 処理効率のためにグループ化
grouped = filtered_df.groupby("horse_id")

# インデックスでループ（未来データを使わないように処理）
for idx, row in filtered_df.iterrows():
    horse_id = row["horse_id"]
    race_date = row["date"]

    # 過去のレースの horse_type のみを取得
    past_types = grouped.get_group(horse_id)
    past_types = past_types[past_types["date"] < race_date]["horse_type"]

    if past_types.empty:
        analy_horse_types.append(0)
    else:
        analy_horse_types.append(past_types.mode().iloc[0])  # 最頻値（脚質傾向）を使用

# 新しい列として追加
filtered_df["analy_horse_type"] = analy_horse_types

# 元の merge_df にマージ（日時・horse_idが一致するもの）
merge_df = merge_df.merge(
    filtered_df[["race_id", "horse_id", "analy_horse_type"]],
    on=["race_id", "horse_id"],
    how="left"
)

# 未定義の馬（出走経験なし）に 0 を補完
merge_df["analy_horse_type"] = merge_df["analy_horse_type"].fillna(0).astype(int)

merge_df.loc[merge_df["analy_horse_type"] != 0, ["analy_horse_type", "horse_type"]].head(20)

merge_df.to_csv(parent_dir + "/merge_df.csv", index=False)
```
